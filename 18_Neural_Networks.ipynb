```python
from sklearn.preprocessing import MinMaxScaler

# Separate the features (x) and the target (y)
x = df.drop('letter', axis=1)
y = df['letter']

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the features
x_scaled = scaler.fit_transform(x)

# Create a new DataFrame with the scaled features
df_scaled = pd.DataFrame(x_scaled, columns=x.columns)

# Add the target variable back to the scaled DataFrame
df_scaled['letter'] = y

# Display the first few rows of the scaled DataFrame
display(df_scaled.head())
```

# **Task 2 : Model Implementation**

**Divide the dataset into training and test sets.**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

**Construct a basic ANN model using your chosen high-level neural network library. Ensure your model includes at least one hidden layer.**

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Encode the target variable
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# One hot encode the target variable
y_train_categorical = to_categorical(y_train_encoded)
y_test_categorical = to_categorical(y_test_encoded)

ann=Sequential()
ann.add(Dense(128,input_dim=x_train.shape[1],activation='relu'))
ann.add(Dense(64,activation='relu'))
ann.add(Dense(y_train_categorical.shape[1],activation='softmax'))

ann.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

**Train your model on the training set and then use it to make predictions on the test set.**

```python
ann.fit(x_train,y_train_categorical,epochs=50,batch_size=100,validation_data=(x_test,y_test_categorical))

prd=ann.predict(x_test)
prd_data=[]
for i in prd:
  prd_data.append(np.argmax(i))
```

# **Task 3 : Hyperparameter Tuning**

**Modify various hyperparameters, such as the number of hidden layers, neurons per hidden layer, activation functions, and learning rate, to observe their impact on model performance.**

```python
def build_and_train_model(num_hidden_layers, neurons_per_layer, activation_function, learning_rate):

    model = Sequential()
    model.add(Dense(neurons_per_layer[0], input_dim=x_train.shape[1], activation=activation_function))
    for i in range(1, num_hidden_layers):
        model.add(Dense(neurons_per_layer[i], activation=activation_function))
    model.add(Dense(y_train_categorical.shape[1], activation='softmax'))

    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    history = model.fit(x_train, y_train_categorical, epochs=50, batch_size=100, validation_data=(x_test, y_test_categorical), verbose=0)

    return model, history

# Function with different hyperparameters
print("Training model with 1 hidden layer, 64 neurons, relu activation, and learning rate 0.001")
model1, history1 = build_and_train_model(num_hidden_layers=1, neurons_per_layer=[64], activation_function='relu', learning_rate=0.001)
print(f"Model 1 final validation accuracy: {history1.history['val_accuracy'][-1]:.4f}")

print("\nTraining model with 2 hidden layers, 128 and 64 neurons, relu activation, and learning rate 0.001")
model2, history2 = build_and_train_model(num_hidden_layers=2, neurons_per_layer=[128, 64], activation_function='relu', learning_rate=0.001)
print(f"Model 2 final validation accuracy: {history2.history['val_accuracy'][-1]:.4f}")

print("\nTraining model with 2 hidden layers, 128 and 64 neurons, tanh activation, and learning rate 0.001")
model3, history3 = build_and_train_model(num_hidden_layers=2, neurons_per_layer=[128, 64], activation_function='tanh', learning_rate=0.001)
print(f"Model 3 final validation accuracy: {history3.history['val_accuracy'][-1]:.4f}")

print("\nTraining model with 2 hidden layers, 128 and 64 neurons, relu activation, and learning rate 0.01")
model4, history4 = build_and_train_model(num_hidden_layers=2, neurons_per_layer=[128, 64], activation_function='relu', learning_rate=0.01)
print(f"Model 4 final validation accuracy: {history4.history['val_accuracy'][-1]:.4f}")
```

**Adopt a structured approach like grid search or random search for hyperparameter tuning, documenting your methodology thoroughly.**

```python
from sklearn.model_selection import ParameterSampler
from scipy.stats import randint, uniform

# Define the hyperparameter search space
param_distributions = {
    'num_hidden_layers': randint(1, 4),  # Number of hidden layers between 1 and 3
    'neurons_per_layer': [[randint.rvs(32, 257, size=1)[0] for _ in range(n)] for n in range(1, 4)], # Neurons per layer between 32 and 256
    'activation_function': ['relu', 'tanh'], # Activation functions
    'learning_rate': uniform(0.0001, 0.01), # Learning rate between 0.0001 and 0.01
}

# Create a parameter sampler for random search
n_iter_search = 20  # Number of parameter settings that are sampled
random_search = ParameterSampler(param_distributions, n_iter=n_iter_search, random_state=42)

best_accuracy = 0
best_params = None
results = []

print(f"Performing random search with {n_iter_search} iterations...")

# Iterate over the sampled hyperparameters and train the model
for i, params in enumerate(random_search):
    print(f"\nIteration {i+1}/{n_iter_search} with parameters: {params}")

    # Adjust neurons_per_layer based on the number of hidden layers
    num_hidden_layers = params['num_hidden_layers']
    neurons_per_layer = [randint.rvs(32, 257, size=1)[0] for _ in range(num_hidden_layers)]
    params['neurons_per_layer'] = neurons_per_layer


    model, history = build_and_train_model(
        num_hidden_layers=params['num_hidden_layers'],
        neurons_per_layer=params['neurons_per_layer'],
        activation_function=params['activation_function'],
        learning_rate=params['learning_rate']
    )

    # Get the final validation accuracy
    final_accuracy = history.history['val_accuracy'][-1]
    print(f"  Validation accuracy: {final_accuracy:.4f}")

    results.append({'params': params, 'accuracy': final_accuracy})

    # Check if this is the best model so far
    if final_accuracy > best_accuracy:
        best_accuracy = final_accuracy
        best_params = params

print("\nRandom search finished.")
print(f"\nBest validation accuracy: {best_accuracy:.4f}")
print(f"Best hyperparameters: {best_params}")

# Optional: Display all results
results_df = pd.DataFrame(results)
display(results_df.sort_values(by='accuracy', ascending=False))
```

# **Task 4 : Evaluation**

**Employ suitable metrics such as accuracy, precision, recall, and F1-score to evaluate your model's performance.**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Assuming 'best_model' is the trained model with the best hyperparameters found
# If you didn't store the best model, you would need to retrain it with best_params
# For now, let's use the last trained model from the random search which is stored in 'model' and 'history'
# However, it's better to retrain the best model explicitly for clarity and correctness.

# Retrain the best model using the best hyperparameters found
print("Retraining the best model with the found hyperparameters...")
best_model, best_history = build_and_train_model(
    num_hidden_layers=best_params['num_hidden_layers'],
    neurons_per_layer=best_params['neurons_per_layer'],
    activation_function=best_params['activation_function'],
    learning_rate=best_params['learning_rate']
)
print("Best model retraining finished.")

# Make predictions with the best model
y_pred_probs = best_model.predict(x_test)
y_pred_encoded = np.argmax(y_pred_probs, axis=1)

# Decode the predicted labels back to original letters
y_test_encoded = label_encoder.transform(y_test)
y_pred_decoded = label_encoder.inverse_transform(y_pred_encoded)

# Evaluate the model using different metrics
accuracy = accuracy_score(y_test_encoded, y_pred_encoded)

# For precision, recall, and F1-score, we need to handle multiclass classification
# 'average' parameter can be 'micro', 'macro', 'weighted', or None
precision = precision_score(y_test_encoded, y_pred_encoded, average='weighted')
recall = recall_score(y_test_encoded, y_pred_encoded, average='weighted')
f1 = f1_score(y_test_encoded, y_pred_encoded, average='weighted')

print(f"\nModel Performance Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```
